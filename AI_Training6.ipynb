CORELATION

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("titanic.csv")
df.head(2)
```

```python
df1=df.select_dtypes(include=['int64','float64'])
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df1.corr(), annot=True, cmap='coolwarm')
plt.show()
```

PIE CHART

```python
import numpy as np
y= np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
y=np.array([35,25,25,15])
l1=["apple","banana","mango","cherry"]
plt.pie(y,labels=l1)
plt.show()
```

```python
y=np.array([35,25,25,15])
l1=["apple","banana","mango","cherry"]
plt.pie(y,labels=l1, startangle=45)
plt.show()
```

```python
y=np.array([35,25,25,15])
l1=["apple","banana","mango","cherry"]
l2=[0.3,0.1,0.1,0.1]
plt.pie(y,labels=l1, explode=l2)
plt.show()
```

```python
y=np.array([35,25,25,15])
l1=["apple","banana","mango","cherry"]
l2=[0.3,0.1,0.1,0.1]
plt.pie(y,labels=l1, explode=l2, shadow=True)
plt.show()
```
