```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Feature_Scaling.csv")
df.head(3)
```

```python
df.describe()
```

```python
df1=df[['Age','Salary']]
```

```python
df1.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(df1)
```

```python
sc.mean_
```

```python
sc.scale_
```

```python
df.describe()
```

```python
df1_sc=sc.transform(df1)
df1_sc
```

```python
#convert into dataframe
df1_sc=pd.DataFrame(df1_sc,columns=['Age','Salary'])
df1_sc.head()
```

MIN MAX SCALER

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
nmc=MinMaxScaler()
nmc.fit(df1)
```

```python
df1_nmc=nmc.transform(df1)
df1_nmc
```

```python
#convert into dataframe
df1_nmc=pd.DataFrame(df1_nmc,columns=['Age','Salary'])
df1_nmc.head()
```

```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.describe()
```

```python
sns.boxplot(df['Fare'],orient='h')
```

```python
sns.boxplot(df['Age'],orient='h')
```

```python
#bi-variable analysis
sns.boxplot(x="Sex",y="Age",data=df)
```

```python
sns.boxplot(x="Survived",y="Age",data=df)
```

SEABORN SCATTER PLOT

```python
sns.scatterplot(x='Age',y='Fare',data=df)
```

```python
sns.scatterplot(x='Age',y='Fare',data=df,hue='Sex')
```

```python
sns.scatterplot(x='Age',y='Fare',data=df,hue='Sex',hue_order=['female','male'])
```
