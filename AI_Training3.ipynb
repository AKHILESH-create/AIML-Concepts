```python
#Count duplicate rows
du=df.duplicated().sum()
print("Total Duplicate rows=", du)
```

```python
#calculate missing values in percentage format
mvp=df.isnull().sum()/df.shape[0]*100
mvp
```

```python
#Selecting only numerical datatype
df1=df.select_dtypes(include=['int64','float64'])
df1.head()
```

```python
#drop column
df1=df1.drop("User ID", axis=1)
df1.shape
```

```python
df1.isnull().sum()
```

```python
import seaborn as sns
sns.heatmap(df1.isnull())
```

```python
b=[var for var in df1.columns
   if df1[var].isnull().sum()>0]
b
```

```python
sns.distplot(df1['Age'])
```

```python
df['Age'].describe()
```

```python
#filling missing value by mean on "Age"
df1['Age']=df1['Age'].fillna(df1['Age'].mean())
df1['Age'].isnull().sum()
```

```python
sns.distplot(df1['EstimatedSalary'])
```

```python
df['EstimatedSalary'].describe()
```

```python
df1['EstimatedSalary']=df1['EstimatedSalary'].fillna(df1['EstimatedSalary'].median())
df1['EstimatedSalary'].isnull().sum()
```

```python
sns.distplot(df1['Purchased'])
```

```python
df['Purchased'].describe()
```

```python
df1['Purchased']=df1['Purchased'].fillna(df1['Purchased'].median())
df1['Purchased'].isnull().sum()
```
