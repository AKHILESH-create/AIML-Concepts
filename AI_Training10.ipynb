```python
import numpy as np
import pandas as pd
import seaborn as sns
df=pd.read_csv("house price prediction.csv")
```

```python
#'Price is our target so we have 107 input so will apply multiple linear regr.
x=df.drop('price',axis=1)
y=df['price']
```

```python
print("Shape of x: ", x.shape)
print("Shape of y: ", y.shape)
```

SPLIT DATA INTO TRAINING AND TESTING

```python
#split data into training and testing
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=66)
```

```python
print("Shape of x_train", x_train.shape)
print("Shape of x_test", x_test.shape)
```

```python
print("Shape of y_train", y_train.shape)
print("Shape of y_test", y_test.shape)
```

FEATURE SCALING

```python
#feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
sc.fit(x_train)
x_train= sc.transform(x_train)
x_test= sc.transform(x_test)
```

ML MODEL TRAINING

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_test,y_test)
```

```python
#predict the value of Flat or home
x_test[0,:]
```

```python
lr.predict([x_test[1,:]])
```

```python
y_test
```

```python
#prediction of test and training data
y_pred=lr.predict(x_test)
x_pred=lr.predict(x_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

#MAE
print("MAE Value: ", mean_absolute_error(y_test, lr.predict(x_test)), mean_squared_error(y_test, lr.predict(x_test)))
```

```python
#MSE
print("MSE Value: ", mean_squared_error(y_test, y_pred))
```

```python
#RMSE
print("RMSE Value: ",np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
#R2
from sklearn.metrics import r2_score
print("R2 Score Value: ", r2_score(y_test, y_pred))
```
