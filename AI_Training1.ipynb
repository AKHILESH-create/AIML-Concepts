SET

```python
s={1,2,-3,-4,5,10,20}
print(type(s))
print(len(s))
```

```python
#adding elements in set
s.add("Python")
print(s)
print(len(s))
```

```python
s.add("ML")
print(s)
print(len(s))
```

```python
s.discard("Python")
print(s)
```

DICTIONARY

```python
dic1={
      "Employee": "Prashant",
      "Id": "800",
      "Salary": 85000
}
print(dic1)
print(type(dic1))
```

```python
print(dic1["Employee"])
```

```python
print(dic1["Salary"])
```

```python
dic1={
      "Employee": "Prashant",
      "Id": "800",
      "Salary": 85000, "Salary":95000
}
print(dic1)
print(type(dic1))
```

```python
dic1["Age"]=35
print(dic1)
```

```python
#update
dic1.update({"Age":37})
print(dic1)
```

```python
dic1.update({"Mobile": 3888888888888})
print(dic1)
```

POP

```python
dic1.pop("Id", None)
print(dic1)
```

```python
dic1 = {'Employee': 'Prashant', 'Salary': 95000, 'Mobile': 3888888888888} # Re-initialize dic1
dic1.popitem()
print(dic1)
```
