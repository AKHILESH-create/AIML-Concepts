```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("student_clustering.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'],df['iq'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss =[]
for i in range(1,11):
  km=KMeans(n_clusters=i)
  km.fit_predict(df)
  wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
x=df.iloc[:,:].values
km=KMeans(n_clusters=4)
# A good model is one with low inertia and low number of clusters(k)
# in our exaple we got low inertia value at cluster 4 so we take n_clusters=4
```

```python
y_pred=km.fit_predict(x)
```

```python
y_pred
```

```python
x[y_pred==3,1]
# 3 indicates cluster number
# 1 indicate colum index "cgpa"
```
