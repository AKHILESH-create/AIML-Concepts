```python
import pandas as pd
print(pd.__version__)
```

```python
#creating series from list
l1= [67,87.80,"India","Japan"]
s1=pd.Series(l1)
print(s1)
print(type(s1))
```

```python
s1=pd.Series([1,2,3,4],index=['a','b','c','d'],dtype=float)
print(s1)
```

```python
#creating series from dictionary
s2=pd.Series({'a':1,"b":2,"c":3})
print(s2)
```
